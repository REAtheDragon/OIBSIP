```python
### Creating my Input and Output variables
```

```python
x = df.drop(columns = ['Time', 'Class'])
y = df['Class']
```

### Importing libraries for Developing the Model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
```

### Splitting the Dataset set into training and testing set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42, stratify = y)
```

### Handling Imbalance in the Dataset

```python
smote = SMOTE(random_state=42)
x_train_res, y_train_res = smote.fit_resample(x_train, y_train)
```

###  Standardizing the columns of the dataset

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_res)
x_test_scaled = scaler.transform(x_test)
```

### Using the Logistic Regression Model

```python
logreg = LogisticRegression(max_iter = 1000, random_state = 42)
logreg.fit(x_train_scaled, y_train_res)
```

```python
y_pred = logreg.predict(x_test_scaled)   
```

### Evaluating the Performance of the Model

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True, fmt = 'd', cmap = 'Blues' )
```

```python
print(classification_report(y_test, y_pred))
```

### Using the Random Forest Model

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42,  n_jobs=-1)
rf.fit(x_train_scaled, y_train_res)
```

```python
y_pred2 = rf.predict(x_test_scaled)
```

### Evaluating the Performance of the Model

```python
cm2 = confusion_matrix(y_test, y_pred2)
```

```python
sns.heatmap(cm2, annot = True, fmt = 'd', cmap = 'Blues' )
```

```python
print(classification_report(y_test, y_pred2))
```

### The Random Forest Model Outperforms the Logistic Regression Model with better performance scores
